```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
s = np.random.dirichlet([0.1] * 10, 10)
print(s)
plt.barh(range(10), s[:, 0])
plt.barh(range(10), s[:, 1], left=s[:, 0], color='g')
plt.barh(range(10), s[:, 2], left=s[:, 0]+s[:, 1], color='r')
plt.title("Lengths of Strings")
```
